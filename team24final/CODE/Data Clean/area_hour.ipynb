```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_hour = pd.read_csv('area_0.csv').fillna(0)
```

```python
segments = pd.unique(data_hour['segmentid'])
plt.figure(figsize=(15,10))
for j, each in enumerate(segments):
    if j < 4:
        continue
    if j == 8:
        break
    plt.subplot(3,2,j+1-4)
    data_segment = data_hour[data_hour['segmentid']==each]
    sum_ratio = np.zeros((24,))
    count = 0
    for i in range(int(len(data_hour)/24)):
        data_each = data_segment.iloc[i*24:i*24+24]
        if np.shape(np.array(data_each['avg_ratio'])) != (0,):
            sum_ratio += np.array(data_each['avg_ratio'])
            count += 1
        plt.plot(np.array(data_each['avg_ratio']), color = 'cyan')
    sum_ratio /= count
    plt.plot(np.array(sum_ratio), color = 'blue', linewidth=3)
```
